```
def read_file(fileName):
    contents = []
    f = open(fileName, 'r')
    for line in f:
      contents.append(line.lower())
    f.close()
    s = '\n'.join(contents)
    
    words = s.split()
    titles = ['mr.', 'mrs.', 'ms.', 'dr.']
    exclamations = ['u.s.a.!', '"u.s.a.!', 'u.s.a.!"', "let's move!"]
    punctuation = ['.', ',', ':', '--', '?', '!', '...']
    interjection = ['(applause.)', '--(applause)--', '[applause]', '--[applause]--', '--[applause]', '[laughter]', '--[laughter]--']
    usa = False

    for w in range(len(words)):
      letters = list(words[w])
      length = len(letters)
      l = 0
      while l < length:
      # for l in range(len(letters)):
        if letters[l] in punctuation:
          if words[w] not in titles and words[w] not in exclamations:
            letters.insert(l, ' ')
            l+=1
        elif letters[l] == '-' and letters[l-1] == '-' and letters[l-2] != ' ':
          letters.insert(l-1, ' ')
          letters.insert(l+2, ' ')
          l+=2
        elif letters[l] == '"':
          letters.remove('"')
        l+=1
        if length != len(letters):
          length = len(letters)
      words[w] = ''.join(letters)   

    result = ' '.join(words)     
    return result

def split(fileName):
  text = read_file(fileName)
  words = text.split(" . ")

  ending_punctuation = ['.', '?', '!']

  with open('train.txt', 'w') as train, open('test.txt', 'w') as test:
    for i in range(len(words)):
      if i%20 != 0 or i == 0:
        train.write(words[i] + ' . \n')
      else:
        test.write(words[i] + ' . \n')

# print(read_file('state_of_the_union_edited.txt'))
split('state_of_the_union_edited.txt')
```
